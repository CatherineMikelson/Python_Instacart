## 2. Busiest Times

### a. Busiest Day of the Week

```python
# create chart
bar_dow = ords_prods_cust['orders_day_of_the_week'].value_counts().sort_index().plot.bar()
bar_dow.set_ylabel('Count of Orders (in millions)')
bar_dow.set_xlabel('Day of the Week')
bar_dow.set_title('Orders by Day of the Week')
```

```python
#save chart as bar_dow.png
bar_dow.figure.savefig(os.path.join(path_git, '04 Analysis', 'Visualizations', 'bar_dow.png'))
```

### b. Busiest Hours of the Day

```python
# create chart
bar_hour = ords_prods_cust['order_hour_of_day'].value_counts().sort_index().plot.bar()
bar_hour.set_ylabel('Count of Orders (in millions)')
bar_hour.set_xlabel('Hour of Day')
bar_hour.set_title('Orders by Hour')
```

```python
#save chart as bar_hour.png
bar_hour.figure.savefig(os.path.join(path_git, '04 Analysis', 'Visualizations', 'bar_hour.png'))
```

## 3. Loyalty Status

### a. Brand Loyalty and Frequency Status

```python
# create crosstab
loyaty_freq = pd.crosstab(df_high_act_prof['loyalty_flag'], df_high_act_prof['Frequency_flag'], dropna = False)
```

```python
# display crosstab
loyaty_freq
```

```python
# create bar
loy_freq_bar = loyaty_freq.plot.bar(rot=0, color = ['navy', 'steelblue', 'lightsteelblue'])
loy_freq_bar.legend(title = None)
plt.title('Frequency Status by Loyalty Status')
plt.ylabel(None)
plt.xlabel(None) 
```

```python
# save chart as bar_dow.png
loy_freq_bar.figure.savefig(os.path.join(path_git, '04 Analysis', 'Visualizations', 'loy_freq_bar_g.png'))
```
