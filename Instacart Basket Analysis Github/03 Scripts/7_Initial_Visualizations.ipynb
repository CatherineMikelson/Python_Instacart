## 2. Initial Charts

### a. Bar Chart for Counts of Order DOW

```python
# create chart of value counts for day of the week
ords_prods_cust['orders_day_of_the_week'].value_counts().plot.bar()
```

```python
# add sort_index
bar = ords_prods_cust['orders_day_of_the_week'].value_counts().sort_index().plot.bar()
```

```python
# add color
bar = ords_prods_cust['orders_day_of_the_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# save chart
bar.figure.savefig(os.path.join(path_git, '04 Analysis', 'Visualizations', 'bar_orders_dow.png'))
```

### b. Histogram of Prices

```python
hist = ords_prods_cust['prices'].plot.hist(bins = 25)
```

```python
# save chart
hist.figure.savefig(os.path.join(path_git, '04 Analysis', 'Visualizations', 'hist_prices.png'))
```

### c. Line Chart DOW and Prices

#### Create Sample

```python
# create a list holding True/False values to the test np.random.rand() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust)) <= 0.7
```

```python
# store 70% of the sample in the dataframe big
big = ords_prods_cust[dev]
```

```python
# store 30% of the sample in the dataframe small
small = ords_prods_cust[~dev]
```

```python
# check data set size
len(ords_prods_cust)
```

```python
# check sample sizes
len(big) + len(small)
```

#### Line Chart

```python
# create order/prices data set with small
df_2 = small[['orders_day_of_the_week', 'prices']]
```

```python
# line chart for day of the week/prices
line = sns.lineplot(data = df_2, x = 'orders_day_of_the_week', y = 'prices')
```

```python
# save line chart as line_prices_dow.png
line.figure.savefig(os.path.join(path_git, '04 Analysis', 'Visualizations', 'line_prices_dow.png'))
```

### d. Histogram  for order_hour_of_day

```python
# create histogram from order_hour_of_day
hist_order_hour = ords_prods_cust['order_hour_of_day'].plot.hist(bins = 48)
```

### The histogram shows that the busiest time of day is midday with orders coming in fairly steadily from 9am - 4pm with 10am as the peak.

## 3. Exploratory Analysis Customer Demographic

### a. Relationship Between Age and Number of Dependants

```python
# create age/number of dependants sample
df_age_dep = small[['age', 'number_of_dependants']]
```

```python
# Line chart for age/number of dependants
line_age_dep = sns.lineplot(data = df_age_dep, x = 'age', y = 'number_of_dependants')
```

Observation: This chart is all over the place with scattered highs and lows.

### b. Connection Between Age and Income

```python
# create age/income sample
df_age_income = small[['age', 'income']]
```

```python
df_age_income.head(5)
```

```python
# create scatter plot for age/income
scatterplot_age_income = sns.scatterplot(x = 'age', y = 'income', data = df_age_income)
```

Observation: After age 40, there is a considerable increase in income.  We see that both the densest points (lower income) and sparser points (higher income) expand upward to see increased income.
