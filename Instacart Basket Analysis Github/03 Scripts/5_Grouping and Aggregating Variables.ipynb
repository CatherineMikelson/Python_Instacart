## 3. Create Loyalty Flag

### a. Create Loyalty Flag

```python
# create max number - the number of orders by user_id
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check output
ords_prods_merged.head(10)
```

```python
#create loyalty flag
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] >10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check output
ords_prods_merged.head()
```

```python
# check value counts
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

### b. Spending Habits Based on Loyalty

Note: Is there a difference between the spending habits of customers based on Loyalty Status?

```python
# basic statistical analysis of loyalty flag by prices
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# calculate loyalty percentage
ords_prods_merged['loyalty_flag'].value_counts(normalize = True)
```

## 4. Create Spending Flag

### a. Create Activity Flag

```python
# create average_price - the average price by user
ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check output
ords_prods_merged.head(10)
```

```python
# round the average price to the 2nd decimal place
ords_prods_merged['average_price'] = ords_prods_merged['average_price'].round(2)
```

```python
# check output
ords_prods_merged.head(50)
```

### b. Create Spending Flag Variable Based on Each Customer's Spending

Note: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.  If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

```python
# create high spender flag
ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merged.loc[ords_prods_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check output
ords_prods_merged.head()
```

```python
# check value counts
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

## 5. Create Frequency Flag

If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer.\
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer.\
If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer.

### a. Create Variable for Median Order Frequency

```python
# create median frquency in days since last order
ords_prods_merged['median_ord_freq'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check output
ords_prods_merged.head()
```

### b. Create Frequency Flag from med_ord_freq

```python
# create frquency flag
ords_prods_merged.loc[ords_prods_merged['median_ord_freq'] > 20, 'Frequency_flag'] = 'Non-frequent customer'
ords_prods_merged.loc[(ords_prods_merged['median_ord_freq'] > 10) & (ords_prods_merged['median_ord_freq'] <= 20), 'Frequency_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['median_ord_freq'] <= 10, 'Frequency_flag'] = 'Frequent customer'
```

```python
# check output
ords_prods_merged.head()
```
